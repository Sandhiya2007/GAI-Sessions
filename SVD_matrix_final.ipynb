```python
import numpy as np

A = [[4,1,1.6,3.4,4.1,3.9,1.9,1.3,1.7],
[2,3,1.8,3.2,3.3,3.7,2.7,1.9,2.1],
[-2,2,0.2,-0.2,-0.8,-0.2,0.8,0.6,0.4],
[7,5,4.1,7.9,8.8,9.1,5.6,3.9,4.6],
[-1,6,2.1,2.9,2.1,3.4,3.9,2.8,2.7],
[2,7,3.4,5.6,5.3,6.5,5.5,3.9,4.1],
[4,1,1.6,3.4,4.1,3.9,1.9,1.3,1.7],
[5,3,2.7,5.3,6,6.1,3.6,2.5,3],
[6,-4,0.2,1.8,3.4,2,-1,-0.8,-0.2],
[1,7,3.1,4.9,4.4,5.7,5.2,3.7,3.8]]

U1, S1, VT1 = np.linalg.svd(A,full_matrices=False) #to get the correct dimensions
S = S1
S1 = np.diag(S1)
App1 = U1 @ S1 @ VT1
print(App1)
print('U1 : ', U1)
print('S1 : ', S1)
print('VT1 : ', VT1)
```

```python
u1 = U1[:, 0:1]
v1 = VT1[0:1,:]
res1 = S[0]*(u1 @ v1)
print(res1)
```
